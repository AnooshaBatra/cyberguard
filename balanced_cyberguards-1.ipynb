```python
import pandas as pd
import re, csv, nltk, string
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

data=pd.read_csv("dataset_balance.csv")
label_bullying=data['label_bullying']
text = data['text_message']
list_text=text.tolist()
```

```python
data
```

```python
type(label_bullying)
import numpy as np
label_bullying=np.array(label_bullying)
type(label_bullying)
```

```python
#lowercase
list_text= [text.lower() for text in list_text]
print(list_text[4])
#number
list_text=[re.sub(r'\d+','',text) for text in list_text]
print(list_text[0])
#remove symbols [!â€#$%&â€™()*+,-./:;<=>?@[\]^_`{|}~]
print(list_text[279])
list_text= [text.translate(str.maketrans('','',string.punctuation)) for text in list_text]
print(list_text[279])
#remove single characters ?
print(list_text[7])
list_text=[re.sub(r'\s+[a-zA-Z]\s+',' ',text) for text in list_text]
print(list_text[7])
#remove single characters
print(list_text[1])
list_text=[re.sub(r'\b[a-zA-Z]\b',' ',text) for text in list_text]
print(list_text[1])
#remove double characters
print(list_text[1])
list_text=[re.sub(r'\b[a-zA-Z][a-zA-Z]\b',' ',text) for text in list_text]
print(list_text[1])
# Substituting multiple spaces with single space
list_text = [re.sub(r'\s+', ' ', text, flags=re.I) for text in list_text]
print(list_text[1])
#remove white space
list_text = [text.strip() for text in list_text]
print(list_text[0])
list_text
```

```python
#tokenize
tokenizer = nltk.tokenize.TreebankWordTokenizer()
list_tokens = [tokenizer.tokenize(text) for text in list_text]
print(len(list_tokens))
print(list_tokens)
```

```python
#stopwords
stop_words = set(stopwords.words('english'))  #nltk file
with open("stopwords.txt") as file:
    for line in file:
        stop_words1=line.split(",")  #opensource
      
tmp=[]
tmplist=[]
for text in list_tokens:
    for i in text:
        if (not i in stop_words) and (not i in stop_words1):
            tmp.append(i)
    tmplist.append(tmp)
    tmp=[]
list_tokens=tmplist
print(list_tokens) 
```

```python
#lemmetizing

lemmatizer= nltk.stem.WordNetLemmatizer()
tmp=[]
tmpl=[]
for i in range(len(list_tokens)):
    for word in list_tokens[i]:
        tmp.append( lemmatizer.lemmatize(word,'n'))
    tmpl.append(tmp)
    tmp=[]
list_tokens=tmpl
#list_tokens
                   
                   
list_tokens=[" ".join(lemmatizer.lemmatize(word,'v') for word in list_tokens[i]) for i in range(len(list_tokens))]

#print(type(list_tokens))
print(list_tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect= CountVectorizer()
x_count= count_vect.fit_transform(list_tokens)
x_count.shape
```

```python
#Tfidf
from sklearn.feature_extraction.text import TfidfTransformer  
tfidfcon = TfidfTransformer()
xtfidf=tfidfcon.fit_transform(x_count)


xtfidf.shape
```

```python
#Tfidf
from sklearn.feature_extraction.text import TfidfVectorizer  
tfidfconverter = TfidfVectorizer(max_features=15000, min_df=3, max_df=0.7) 
corpus=list_tokens
x = tfidfconverter.fit_transform(corpus)

feature_names=tfidfconverter.get_feature_names()
feature_vectors=x.toarray()
print(len(feature_names))
print(feature_vectors)
x.shape
```

```python
#saving as vectorizer pickle
import pickle
with open('vectorizer.pickle', 'wb') as picklefile:  
    pickle.dump(tfidfconverter,picklefile)
print("done")
```

```python
#train test divide
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(feature_vectors, label_bullying, test_size=0.3, random_state=0) 
print("done divide!")
x_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(x_train, y_train)
print("done Successfully")
```

```python
#naivebayes
predicted = clf.predict(x_test)
print("done")
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,predicted))  
print(classification_report(y_test,predicted))  
print("Acuracy: ",accuracy_score(y_test, predicted)*100) 
```

```python
#svm

predicted_svm = svmclf.predict(x_test)
print("done")
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,predicted_svm))  
print(classification_report(y_test,predicted_svm))  
print("Acuracy: ",accuracy_score(y_test, predicted_svm)*100) 
```
